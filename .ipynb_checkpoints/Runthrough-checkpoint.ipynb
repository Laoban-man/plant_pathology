# Project presentation

Previous attempts at classifying plant diseases from images have yield positive results. This Kaggle challenge addresses the issue of visual variations in symptoms by providing a much richer training dataset which exposes a ML algorithm to variations in natural and image capturing environments including but not limited to leaf color and leaf morphology, the age of infected tissues, non-uniform image background, and different light illumination during imaging.

To address this image classification problem, we can further augment the training dataset by applying a number of image modifcations, such as playing on brigthness, angle, noise, and contrast. The next step is building a neural network and deciding whether to benefit from transfer learning or training a new model from scratch.

# Optional data augmentation

A possible improvement to the results can be made by augmenting the data. This entails bringing different modifications to the training images (scale, rotation, cropping, ...)

```python
import solt
import solt.transforms as slt

h,w,c = img.shape

stream = solt.Stream([
    slt.Rotate(angle_range=(-90, 90), p=1, padding='r'),
    slt.Flip(axis=1, p=0.5),
    slt.Scale(range_x=(0.8, 1.3), padding='r', range_y=(0.8, 1.3), same=False, p=0.5),
    slt.CvtColor('rgb2gs', keep_dim=True, p=0.2),
    slt.HSV((0, 10), (0, 10), (0, 10)),
], ignore_fast_mode=True)

fig = plt.figure(figsize=(16,16))
n_augs = 6


random.seed(42)
for i in range(n_augs):
    img_aug = stream({'image': img}, return_torch=False, ).data[0].squeeze()

    ax = fig.add_subplot(1,n_augs,i+1)
    if i == 0:
        ax.imshow(img)
    else:
        ax.imshow(img_aug)
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
import Augmentor
p = Augmentor.Pipeline(source_directory=train_files_path,output_directory="../output")
p.rotate90(probability=0.5)
p.rotate270(probability=0.5)
p.flip_left_right(probability=0.8)
p.flip_top_bottom(probability=0.3)
p.crop_random(probability=1, percentage_area=0.5)
p.resize(probability=1.0, width=120, height=120)

p.sample(100)
```

## Model building

```python
from dpn_network import *
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.preprocessing import OneHotEncoder
sample_file_path="../data/plant_pathologies/output"
sample_files=glob(sample_file_path+"/*.jpg")
X=[]
y=[]
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(augmented_files_labelling[['labels']]).toarray())# merge with main df bridge_df on key values
augmented_files_labelling = augmented_files_labelling.join(enc_df)
for file in sample_files:
    X.append(plt.imread(file))
    y.append(augmented_files_labelling[augmented_files_labelling["image"]==file[55:75]].iloc[0,2:])

X=np.asarray(X).astype(np.float32)
y=np.asarray(y).astype(np.int32)
```

```python
X_reshaped.shape
```

```python
from sklearn.model_selection import train_test_split as tts
from imblearn.over_sampling import RandomOverSampler
X_reshaped=X.reshape(X.shape[0],X.shape[1]*X.shape[2]*X.shape[3])
ros=RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X_reshaped,y)
#sm = SMOTE(random_state=42)
#X_res, y_res = sm.fit_resample(X_reshaped,y)

X_res=X_res.reshape(X_res.shape[0],120,120,3)
X_train,X_test,y_train,y_test=tts(X_res,y_res,train_size=0.8)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
```

```python
model = DPN92(input_shape=(120, 120, 3),classes=12)
```

```python
model.compile(optimizer='adam',metrics = ['accuracy'],loss = 'categorical_crossentropy')
model.fit(train_datagen.flow(X_train,y_train,batch_size = 64),validation_data=(X_test,y_test),epochs = 40)
```

```python
y_pred = model.predict(X_test[0:10,:,:,:])
print(y_pred)
print(y_test[0:10,:])
```
